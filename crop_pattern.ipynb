```python
import pandas as pd

# Read your file
df = pd.read_csv("crop.csv")

# Extract unique Crop–Season pairs
unique_df = df[['Crop', 'Season']].drop_duplicates().reset_index(drop=True)

# Define mapping for each season
season_mapping = {
    'Kharif': {
        'Sowing Period': ['June', 'July'],
        'Harvesting Period': ['October', 'November']
    },
    'Rabi': {
        'Sowing Period': ['October', 'November', 'December'],
        'Harvesting Period': ['March', 'April']
    },
    'Zaid': {
        'Sowing Period': ['March', 'April'],
        'Harvesting Period': ['June', 'July']
    },
    'Summer': {
        'Sowing Period': ['February', 'March'],
        'Harvesting Period': ['May', 'June']
    },
    'Winter': {
        'Sowing Period': ['October', 'November'],
        'Harvesting Period': ['February', 'March']
    },
    'Autumn': {
        'Sowing Period': ['September', 'October'],
        'Harvesting Period': ['November', 'December']
    },
    'Whole Year': {
        'Sowing Period': ['January', 'February', 'March', 'April', 'May', 'June', 
                          'July', 'August', 'September', 'October', 'November', 'December'],
        'Harvesting Period': ['January', 'February', 'March', 'April', 'May', 'June', 
                              'July', 'August', 'September', 'October', 'November', 'December']
    }
}

# Map new columns based on Season
unique_df['Sowing Period'] = unique_df['Season'].map(lambda s: season_mapping.get(s, {}).get('Sowing Period', []))
unique_df['Harvesting Period'] = unique_df['Season'].map(lambda s: season_mapping.get(s, {}).get('Harvesting Period', []))

# Save the cleaned dataset
unique_df.to_csv("crop_pattern.csv", index=False)

print("✅ CSV file 'crop_pattern.csv' created successfully with columns: Crop, Season, Sowing Period, Harvesting Period")
```
