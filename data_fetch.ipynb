```python
# Reload the data to start fresh
df = pd.read_csv('India Agriculture Crop Production.csv')

print("Original Year column:")
print(df['Year'].unique()[:10])

# Keep only the text after "-" in the Year column
df['Year'] = df['Year'].str.split('-').str[1]

print("\nAfter extracting text after '-':")
print(df['Year'].unique()[:10])

# Add "20" before the values
df['Year'] = '20' + df['Year']

print("\nAfter adding '20' prefix:")
print(df['Year'].unique()[:10])

# Display the updated dataframe
df.head(10)
```

```python
# Convert Year column to datetime type
print("Before conversion:")
print(f"Data type: {df['Year'].dtype}")
print(f"Sample values: {df['Year'].unique()[:5]}")

# Convert Year column to datetime (will create dates as January 1st of each year)
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

print("\nAfter datetime conversion:")
print(f"Data type: {df['Year'].dtype}")
print(f"Sample values: {df['Year'].unique()[:10]}")

# Display the updated dataframe
print("\nDataframe info:")
print(df.info())
print("\n")
df.head(10)
```
