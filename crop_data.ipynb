```python
# Reload the data to start fresh
df = pd.read_csv('India Agriculture Crop Production.csv')

print("Original Year column:")
print(df['Year'].unique()[:10])

# Keep only the text after "-" in the Year column
df['Year'] = df['Year'].str.split('-').str[1]

print("\nAfter extracting text after '-':")
print(df['Year'].unique()[:10])

# Add "20" before the values
df['Year'] = '20' + df['Year']

print("\nAfter adding '20' prefix:")
print(df['Year'].unique()[:10])

# Display the updated dataframe
df.head(10)
```

```python
# Convert Year column to datetime type
print("Before conversion:")
print(f"Data type: {df['Year'].dtype}")
print(f"Sample values: {df['Year'].unique()[:5]}")

# Convert Year column to datetime (will create dates as January 1st of each year)
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

print("\nAfter datetime conversion:")
print(f"Data type: {df['Year'].dtype}")
print(f"Sample values: {df['Year'].unique()[:10]}")

# Display the updated dataframe
print("\nDataframe info:")
print(df.info())
print("\n")
df.head(10)
```

```python
# Convert columns to appropriate data types
print("Data types BEFORE conversion:")
print(df.dtypes)
print("\n" + "="*80 + "\n")

# Convert to string/text data type
df['State'] = df['State'].astype(str)
df['District'] = df['District'].astype(str)
df['Crop'] = df['Crop'].astype(str)
df['Season'] = df['Season'].astype(str)
df['Area Units'] = df['Area Units'].astype(str)
df['Production Units'] = df['Production Units'].astype(str)

# Convert to float data type
df['Production'] = df['Production'].astype(float)
df['Area'] = df['Area'].astype(float)
df['Yield'] = df['Yield'].astype(float)

print("Data types AFTER conversion:")
print(df.dtypes)
print("\n" + "="*80 + "\n")

print("Summary:")
print(f"✓ State, District, Crop, Season, Area Units, Production Units → string (object)")
print(f"✓ Production, Area, Yield → float64")
print(f"✓ Year → datetime64[ns]")
print("\n")

# Display sample data
df.head(10)
```

```python
# Save the processed dataset to CSV file
print("Saving dataset to 'crop.csv'...")
print(f"Total rows to save: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"Columns: {list(df.columns)}")
print("\n")

# Save to CSV
df.to_csv('crop.csv', index=False)

print("✓ Dataset successfully saved to 'crop.csv'")
print(f"✓ File saved in: c:\\Users\\dishi\\Downloads\\crop\\crop.csv")
print(f"✓ Rows saved: {len(df)}")
print(f"✓ Year range: {df['Year'].dt.year.min()} to {df['Year'].dt.year.max()}")
```

```python
# Convert Production values to Tonnes based on Production Units
print("="*80)
print("CONVERTING PRODUCTION TO TONNES")
print("="*80)

print("\nBefore conversion:")
print(f"Total rows: {len(b)}")
print(f"Production Units distribution:")
print(b['Production Units'].value_counts())
print("\n" + "="*80 + "\n")

# Conversion factors (approximate standard conversions)
# 1 Bale of cotton ≈ 170 kg = 0.17 tonnes
# 1,000 Nuts (coconut) ≈ 1.4 tonnes (average coconut weight ~1.4 kg)

# Convert Bales to Tonnes
bales_mask = b['Production Units'] == 'Bales'
if bales_mask.sum() > 0:
    print(f"Converting {bales_mask.sum()} rows from Bales to Tonnes...")
    b.loc[bales_mask, 'Production'] = b.loc[bales_mask, 'Production'] * 0.17
    b.loc[bales_mask, 'Production Units'] = 'Tonnes'

# Convert Nuts to Tonnes (assuming per 1000 nuts)
nuts_mask = b['Production Units'] == 'Nuts'
if nuts_mask.sum() > 0:
    print(f"Converting {nuts_mask.sum()} rows from Nuts to Tonnes...")
    # Divide by 1000 to get thousands of nuts, then multiply by 1.4 kg/nut average, then divide by 1000 to get tonnes
    b.loc[nuts_mask, 'Production'] = (b.loc[nuts_mask, 'Production'] / 1000) * 1.4
    b.loc[nuts_mask, 'Production Units'] = 'Tonnes'

print("\n" + "="*80)
print("\nAfter conversion:")
print(f"Production Units distribution:")
print(b['Production Units'].value_counts())
print("\n" + "="*80)

print("\n✓ All production values converted to Tonnes")
print(f"✓ Conversion factors used: Bales=0.17t, Nuts=1.4kg/1000")

# Display sample
b.head(10)
```

```python
# Get unique values in Area Units column
if 'Area Units' in b.columns:
    unique_area_units = b['Area Units'].dropna().unique()
    print(f"Unique values in 'Area Units' column ({len(unique_area_units)}):")
    print(sorted(unique_area_units))
    print("\n" + "="*80)
    print("\nValue counts:")
    print(b['Area Units'].value_counts())
else:
    print("No 'Area Units' column found in the dataframe.")
```

```python
# Recalculate Yield for all rows where Production was converted to Tonnes
print("="*80)
print("RECALCULATING YIELD AFTER PRODUCTION UNIT CONVERSION")
print("="*80)

# Identify rows where Production Units is now 'Tonnes' (these were converted)
# Recalculate Yield = Production / Area for all rows
print(f"\nRecalculating Yield for all rows...")
print(f"Formula: Yield = Production / Area")

# Store original yield for comparison
original_yield_sample = b[['State', 'District', 'Crop', 'Year', 'Production', 'Area', 'Yield']].head(10).copy()

# Recalculate Yield
b['Yield'] = b['Production'] / b['Area']

# Display comparison
print("\n" + "="*80)
print("\nSample of recalculated data:")
print(b[['State', 'District', 'Crop', 'Year', 'Production', 'Area', 'Yield']].head(10))

print("\n" + "="*80)
print("\n✓ Yield recalculated for all rows")
print(f"✓ New Yield = Production (Tonnes) / Area (Hectare)")

# Check for any NaN or infinite values
nan_count = b['Yield'].isna().sum()
inf_count = (b['Yield'] == float('inf')).sum()
print(f"\nData quality check:")
print(f"  - NaN values in Yield: {nan_count}")
print(f"  - Infinite values in Yield: {inf_count}")

b.head(10)
```
