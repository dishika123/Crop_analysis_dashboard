```python
import pandas as pd

# Read the two datasets
a = pd.read_csv("crop.csv")
b = pd.read_csv("rain.csv")

# --- Clean the 'State' column to avoid hidden mismatches ---
a['State'] = a['State'].str.strip().str.title()
b['State'] = b['State'].str.strip().str.title()

# --- Inner join on both State and Year ---
merged_df = pd.merge(a, b, on=['State', 'Year'], how='inner')

# --- Optional sanity check ---
print("✅ Merge complete.")
print("Merged shape:", merged_df.shape)
print("Columns:", merged_df.columns.tolist())

# --- Save the merged dataset ---
merged_df.to_csv("merge.csv", index=False)
print("File saved as merged_crop_rain.csv")
```
